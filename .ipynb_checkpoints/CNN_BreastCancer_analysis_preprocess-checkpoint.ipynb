# Project Breast Histopathology Images

In this project we present our Convolutional Neural Network for automatic detection and visual analysis of invasive ductal carcinoma (IDC) tissue regions from whole slide images (WSI) of breast cancer. We use a [dataset from kaggle](https://www.kaggle.com/paultimothymooney/breast-histopathology-images), where the original images, belonging to different patients, have already been preprocessed and divided into small image patches of size 50x50. The patches are classified with either 1 for containing IDC or with 0 for not containing IDC.

Precise delineation of IDC in WSI is crucial to the subsequent estimation of grading tumor aggressiveness and predicting patient outcome. Deep Learning approaches such as the appliance of ConV Neural Networks offer a particularly interesting method for such kinds of problems, especially if a large number of training images is at present.

The dataset from kaggle contains images from 279 patients, divided into subfolders for each patient. Each patient id has two subfolders '0' and '1' containing either image patches belonging to that patient without IDC on them or with IDC.

We divided our project into two notebooks.

This first notebook will apply some interesting exploration and necessary preprocessing on the data, present a first analysis on what we can see already. In the second notebook, named [CNN_BreastCancer_pytorch_3layers](https://github.com/LauresMaures/BreastCancer/blob/master/CNN_BreastCancer_pytorch_3layers.ipynb) we will present our model design for a Convolutional Neural Network, train our model and evaluate the results. We used the following paper from [Cruz-Roa et al. (2014)](https://engineering.case.edu/centers/ccipd/sites/ccipd.case.edu/files/Automatic_detection_of_invasive_ductal_carcinoma_in_whole.pdf) as a guideline. These sources were also particularly helpful and inspiring to us: [pytorch_kfkd](https://gist.github.com/ptrblck/aca3a06274f0bc2244b1401451389993), [Sasank Chilamkurthy](https://pytorch.org/tutorials/beginner/transfer_learning_tutorial.html), [Allunia](https://www.kaggle.com/allunia/breastcancer), [Mazumder et al. (2019)](https://mc.ai/skin-cancer-image-classification-an-educational-guide/), [Bukun](https://www.kaggle.com/ambarish/breast-histopathology-pytorch).

```python
# create empty data frame
data = pd.DataFrame(index=np.arange(0, noAllPatches), columns=["patient_id", "image_path", "label"])

# define base path
basePath = "data/breast-histopathology-images/IDC_regular_ps50_idx5/"

# fill data frame
def fillDataFrame(df, path, patients):
    i = 0
    for j in range(len(patients)):
        patientId = patients[j]
        patientPath = path + str(patientId)
        for label in [0,1]:
            classPath = patientPath + "/" + str(label) + "/"
            subfiles = listdir(classPath)
            for k in range(len(subfiles)):
                imagePath = subfiles[k]
                df.iloc[i]["image_path"] = classPath + imagePath
                df.iloc[i]["label"] = label
                df.iloc[i]["patient_id"] = patientId
                i += 1  
                
    return df
```

### Reconstruction of tissue samples as a whole

As already mentioned, the images from our data set are patches of size 50x50 that have been taken systematically from larger images, the patient's at least in parts infected tissue. The position of every patch in the tissue is saved in the file name by their x and y coordinates. Hence, the tissue per patient can be (almost) fully recomposed. 

In the following section, first, we recompose the labels as a dot map. Then, we reconstruct the nearly complete picture from the original images. 

The visualisations are not complete as they display gaps, as we will see further down. The reason for this phenomenon is that patches with mostly fatty tissue or slide background were discarded, as [Cruz-Roa et al. (2014)](https://engineering.case.edu/centers/ccipd/sites/ccipd.case.edu/files/Automatic_detection_of_invasive_ductal_carcinoma_in_whole.pdf) state.

#### Extract coordinates from image patches 

```python
# extract x and y coordinates of each patch in tissue sample images and append to df
def extractCoords(df):
    coord = df.image_path.str.rsplit("_", n=4, expand=True)
    coord = coord.drop([0, 1, 4], axis=1)
    coord = coord.rename({2: "x", 3: "y"}, axis=1)
    coord.loc[:, "x"] = coord.loc[:,"x"].str.replace("x", "", case=False).astype(np.int)
    coord.loc[:, "y"] = coord.loc[:,"y"].str.replace("y", "", case=False).astype(np.int)
    df.loc[:, "x"] = coord.x.values
    df.loc[:, "y"] = coord.y.values
    return df
```

```python
# apply extract coordinates
dfWithCoords = extractCoords(loaded_df)
```

```python
# function for extracting data frame for one patient through patient id
def getPatientDf(patient, coord_df):
    patient_df = coord_df.loc[coord_df["patient_id"] == patient]
    return patient_df
```

### Visualise full tissue image by recomposing the original patches

```python
# function to reconstruct full tissue for one patient using the x and y coordinates

def reconstructFullTissue(patient_id, coord_df):
    patient_label_df = getPatientDf(patient_id, coord_df)
    max_coords = [patient_label_df.y.max()-1, patient_label_df.x.max()-1]
    
    grid = 255*np.ones(shape = (max_coords[0] + 50, max_coords[1] + 50, 3)).astype(np.uint8)
    mask = 255*np.ones(shape = (max_coords[0] + 50, max_coords[1] + 50, 3)).astype(np.uint8) 
    
    mask_proba = np.zeros(shape = (max_coords[0] + 50, max_coords[1] + 50, 1)).astype(np.float)
    
    broken_patches = []
    for i in range(len(patient_label_df)):
        try:
            image = cv2.imread(patient_label_df.image_path.values[i])
            
            label = patient_label_df.label.values[i]
            
            x_coord = np.int(patient_label_df.x.values[i])
            y_coord = np.int(patient_label_df.y.values[i])
            x_begin = x_coord - 1
            y_begin = y_coord - 1
            x_end = x_begin + 50
            y_end = y_begin + 50

            grid[y_begin:y_end, x_begin:x_end] = image
            if label == 1:
                mask[y_begin:y_end, x_begin:x_end, 0] = 250
                mask[y_begin:y_end, x_begin:x_end, 1] = 0
                mask[y_begin:y_end, x_begin:x_end, 2] = 0

        except ValueError:
            broken_patches.append(patient_label_df.image_path.values[i])    
    
    return grid, mask, broken_patches, mask_proba
```

```python
# function for plotting the full tissue of one patient
def plotFullTissue(patient_id, coord_df):
    
    grid, mask, broken_patches,_ = reconstructFullTissue(patient_id, coord_df)

    fig, ax = plt.subplots(1,2,figsize=(20,10))
    ax[0].imshow(grid, alpha=0.9)
    ax[1].imshow(mask, alpha=0.8)
    ax[1].imshow(grid, alpha=0.7)
    ax[0].grid(False)
    ax[1].grid(False)
    for i in range(2):
        ax[i].set_xlabel("x coordinates")
        ax[i].set_ylabel("y coordinates")
    ax[0].set_title("Tissue sample of patient " + str(patient_id))
    ax[1].set_title("Tissue sample of patient " + str(patient_id) + "\n Patches labeled malign are inked");
    
    # print("broken patches: " + str(broken_patches))
```

## The imbalanced class problem

As we have already seen in one of the statistics shown above, the data set is imbalanced: Class 0 is the majority class containing around 2.5 times more patches than minority class 1. This phenomenon is common in medical data where infected classes typically tend to be smaller than their healthy counterparts. This class imbalance is problematic because during training the healthy class is learned much better than the infected one resulting in a one-way optimization and biased conclusions. For this very reason, we will augment class 1 later on.  

```python
size_class0 = len(loaded_df[loaded_df['label'] == 0]) #198.738
size_class1 = len(loaded_df[loaded_df['label'] == 1]) #78.786 

print(str(size_class0) + " images belong to class 0, while class 1 has a size of just " + str(size_class1))
print("--> difference of " + str(size_class0 - size_class1) + " between the two classes")
print("--> class 0 is ~ " + str("{0:.2f}".format(size_class0/size_class1)) + " times larger than class 1")
```

## Data augmentation

In the following, we will use data augmentation on the training set in order to circumvent the problem of class imbalance that we have outlined before. We will boost the number of image patches with class 1 as it is much smaller than class 0. The aim is to augment the data of class 1 (malign), the minority class, until it reaches the size of majority class 0. 

In the first section we will do some "manual" augmentation and visualize the results to get an idea of what is happening. We will perform image rotation and flipping of original image patches.


```python
# preprocess patch for augmentation
def preprocessPatch(image_path, color_mode):
    patch = cv2.imread(image_path)
    img_as_arr = Image.fromarray(patch, color_mode) #color_mode kann auch weggelassen werden; YUV nehmen??
    return img_as_arr  
```

```python
# augment a single patch: rotate by variations of 90 and mirror
# this function also adds labels so the original image and the augmented counterparts can be visualized
def augmentPatch(preprocessed_image):
    augmented_images = []
    aug_labels = []
    
    # add original patch first for visualization purpose
    augmented_images.append(preprocessed_image)
    aug_labels.append('original')
    
    # rotate original patch:
    rotation_angles = [90, 180, 270]
    for angle in rotation_angles:
        rotated_img = preprocessed_image.rotate(angle)
        augmented_images.append(rotated_img)
        aug_labels.append(angle)

    # mirror the patch:
    mirrored_img = ImageOps.mirror(preprocessed_image)
    augmented_images.append(mirrored_img)
    aug_labels.append('mirrored')
    
    return augmented_images, aug_labels    
```

#### Visualization of chosen set of transformations for augmentation for one example patch

Next, we visualise the transformations for augmentation to give an idea of the process.

```python
# plot augmented set of one patch - visualize original image and its transformed versions
def plotAugmentation(augmented_img, aug_labels):
    plt.figure(1, figsize=(20,20))
    n = 1
    for img in augmented_img:
        plt.subplot(3,3,n)
        plt.imshow(img)
        if(isinstance(aug_labels[n-1], int)):
            plt.title("rotated by " + str(aug_labels[n-1]) + "")
        else:
            plt.title(aug_labels[n-1])
        n +=1

    plt.show()
```

```python
#example patch:
one_img_path = loaded_df.loc[0]['image_path']
one_img_class = loaded_df.loc[0]['label']

# preprocess example patch
img_for_aug = preprocessPatch(one_img_path, color_mode='RGB')

# augment preprocessed example patch
augmented_img1, aug_labels = augmentPatch(img_for_aug)

# plot augmented patch
print("Augmentation example plot for one sample patch showing entire set of transformations")
plotAugmentation(augmented_img1, aug_labels)
```

### Actual augmentation implementation

We decided to generate just two augmentations per patch as we only need to achieve an equal number of patches in class 0 and 1. Tripling the images is more than sufficient to fill up class 1 to the level of class 0.

```python
# patch augmentation function with randomly selected transformations used for our real image augmentation

def finalAugmentPatch(preprocessed_image):
    augmented_images = []
    
    #augment patch randomly (choose between 3 rotation options or mirroring randomly):
    augmentation_options = [90, 180, 270, 'mirror']
    
    for index in range(2):
        rand = random.randint(0,3)
        chosen_aug = augmentation_options[rand]

        if(isinstance(chosen_aug, int)):
            augmented_img = preprocessed_image.rotate(chosen_aug)
            augmented_images.append(augmented_img)
        else:
            augmented_img = ImageOps.mirror(preprocessed_image)
            augmented_images.append(augmented_img)
    
    return augmented_images
```

#### Visualize actual augmentation function

```python
# plot both randomly chosen augmentation transformations of one patch - visualize the two transformed variants only
def plotAugmentation2(augmented_img):
    plt.figure(1, figsize=(20,20))
    n = 1
    for img in augmented_img:
        plt.subplot(3,3,n)
        plt.imshow(img)
        n +=1
    plt.show()
```

#### Create data frame for augmented patches 

```python
# get file names
all_images = listdir(filepath)

# create new data frame
aug_data = pd.DataFrame(all_images, columns=['image_name'])

# show new data frame
aug_data
```

```python
# define functions to extract information about data from file names

# extract patient
def extractPatientId(file_name):
    # split into list
    a = file_name.split('_')
    patient_id = a[0]    
    return patient_id

# extract diagnosed class
def extractLabel(file_name):
    a = file_name.split('_')
    # the label is part of the file name at index 4
    b = a[4]
    # label 0 or 1 is the 5th index of the string --> class1
    label = b[5]    
    return label

# reconstruct full file path
def createImagePath(file_name):
    return filepath + file_name
```

```python
# create new columns for the data frame and fill them with extracted information on patches
aug_data['patient_id'] = aug_data['image_name'].apply(extractPatientId)
aug_data['image_path'] = aug_data['image_name'].apply(createImagePath)
aug_data['label'] = aug_data['image_name'].apply(extractLabel)

# drop obsolete column 'image_name'
aug_data = aug_data.drop(['image_name'], axis=1)
```

```python
# show data frame
aug_data
```

```python
# extract coordinates for data frame of augmented class 1 patches
def extractCoordsForAugData(df):
    coord = df.image_path.str.rsplit("_", n=5, expand=True)
    coord = coord.drop([0, 1, 4, 5], axis=1)
    coord = coord.rename({2: "x", 3: "y"}, axis=1)
    coord.loc[:, "x"] = coord.loc[:,"x"].str.replace("x", "", case=False).astype(np.int)
    coord.loc[:, "y"] = coord.loc[:,"y"].str.replace("y", "", case=False).astype(np.int)
    df.loc[:, "x"] = coord.x.values
    df.loc[:, "y"] = coord.y.values
    return df
```

```python
# apply function from above to get all augmented data with coordinates (position of patch in tissue image)
aug_data_with_coords = extractCoordsForAugData(aug_data)

# number of augmented patches
nr_augmented_data = len(aug_data_with_coords)

# size of training data labeled class 0
len_train_class0 = len(train_df_class0)

# size of training data labeled class 1
len_train_class1 = len(train_df_class1)

# calculate number of missing patches in minority class 1 compared to majority class 0
nr_missing_patches_for_class1 = len_train_class0 - len_train_class1

print("Number of newly augmented data from train set (class 1 only): " + str(nr_augmented_data))
print("Original number of patches in class 1: " + str(len_train_class1))
print("Number for both classes in training set aimed at: " + str(len_train_class0))
print("Number of patches we will take from augmented data to fill up class 1: " + str(nr_missing_patches_for_class1))
```

### Extract right number of patches from augmented data set to fill up class 1 in training set

```python
# in order to reach class balance extract necessary number of 84507 patches from augmented set
aug_samples = aug_data_with_coords.sample(n=nr_missing_patches_for_class1)
aug_samples
```

### Merge the extracted samples from the augmented class 1 data with the rest of the data from the training set and obtain our final training data set

```python
# merge training set with augmented patches
final_train_df = pd.concat([train_df, aug_samples], ignore_index=True)
train_size = len(final_train_df)
print("The size of our final training data set is: " + str(train_size)) # should be 280832
```
